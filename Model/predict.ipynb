```python
def backtest1(data, model, predictors, start=2, step=1):
    all_predictions = []

    seasons = sorted(data["season"].unique())

    for i in range(start, len(seasons), step):
        season = seasons[i]

        train = data[data["season"] < season]
        test = data[data["season"] == season]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)

        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual","prediction"]

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
def calculate_inflator(base_prediction, expected_ppg, std_dev, lower_bound=0.75, upper_bound=1.25):
    # Calculate the difference
    difference = expected_ppg - base_prediction
    
    # Calculate the z-score
    z_score = difference / std_dev
    
    # Calculate the percentile based on the z-score
    percentile = stats.norm.cdf(z_score) * 100
    
    # Map the percentile to the inflator range [lower_bound, upper_bound]
    inflator = np.interp(percentile, [0, 50, 100], [lower_bound, 1, upper_bound])
    
    return inflator
```

```python
def apply_inflator(base_predictions, df_advanced_stats, std_dev, inflator_weight=1):
    # Copy the base_predictions DataFrame and add a new column for adjusted predictions
    adjusted_predictions = base_predictions.copy()
    adjusted_predictions['prediction adj'] = adjusted_predictions['prediction']  # Initialize with base predictions
    
    for idx, row in base_predictions.iterrows():
        team_full = row['Team_FullName']
        season = row['season']
        opponent = row['Team_FullName_Opp']  # Use the team_opp column
        
        previous_season = season - 1 + 1
        
        # Retrieve the offensive rating for the proponent team for the previous season
        team_stats_row = df_advanced_stats[(df_advanced_stats['team'] == team_full) & (df_advanced_stats['season'] == previous_season)]
        if not team_stats_row.empty:
            team_stats_row = team_stats_row.iloc[0]
            team_ortg = team_stats_row['ORtg']
        else:
            team_ortg = None
            print(f"No matching row found for team: {team_full}, previous season: {previous_season}")

        # Retrieve the defensive rating for the opponent team for the previous season
        opponent_stats_row = df_advanced_stats[(df_advanced_stats['team'] == opponent) & (df_advanced_stats['season'] == previous_season)]
        if not opponent_stats_row.empty:
            opponent_stats_row = opponent_stats_row.iloc[0]
            opponent_drtg = opponent_stats_row['DRtg']
            pace = (team_stats_row['Pace'] + opponent_stats_row['Pace'])/2 if team_stats_row is not None else opponent_stats_row['Pace']
        else:
            opponent_drtg = None
            print(f"No matching row found for opponent: {opponent}, previous season: {previous_season}")

        if team_ortg is not None and opponent_drtg is not None:
            # Calculate the expected PPG
            expected_ppg = (pace * ((team_ortg + opponent_drtg) / 2)) / 100
            
            # Calculate inflator
            inflator = calculate_inflator(row['prediction'], expected_ppg, std_dev)
            print(f"Team: {team_full}, Opponent: {opponent}, season: {season}, Team ORtg: {team_ortg}, Opp DRtg: {opponent_drtg}, Pace: {pace}, prediction: {row['prediction']}, expected PPG: {expected_ppg}, inflator: {inflator}")
            
            # Apply inflator to the prediction
            adjusted_prediction = row['prediction'] * inflator
            adjusted_predictions.at[idx, 'prediction adj'] = adjusted_prediction
    
    return adjusted_predictions
```
